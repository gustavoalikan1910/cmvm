```python
import requests

BASE_URL = "https://www.statshub.com/api"
r = requests.get(f"{BASE_URL}/tournament")
tournaments = r.json()["data"]

for country, leagues in tournaments.items():
    for league in leagues:
        if "libert" in league["slug"].lower() or "libert" in league.get("name", "").lower():
            print(league)
```

```python
for country, leagues in tournaments.items():
    for league in leagues:
        if "south-america" in str(league.get("country", "")).lower() or "conmebol" in league.get("name", "").lower():
            print(league)
```

```python
for league in [{"name": "Libertadores", "internalId": 255}, {"name": "Sudamericana", "internalId": 1708}, {"name": "Copa America", "internalId": 1205}]:
    r = requests.get(f"{BASE_URL}/tournament/{league['internalId']}/seasons")
    data = r.json().get("data", [])
    print(f"{league['name']}: {len(data)} seasons")
    for s in data:
        print(f"  {s}")
```
